```python
def gradienteDescendenteP(Theta, g, t, alpha):
    m = len(t)
    
    for _ in range(15000):
        Sum0, Sum1, Sum2 = 0,0,0
        
        for i in range(m):
            hi = Theta[0] + Theta[1]*t[i] + Theta[2]*(t[i])**2
            gi = g[i]
            
            temp = (hi - gi)
            
            Sum0 += temp*1
            Sum1 += temp*t[i]
            Sum2 += temp*(t[i])**2
                
        Theta[0] = Theta[0] - 2*alpha*Sum0/m
        Theta[1] = Theta[1] - 2*alpha*Sum1/m
        Theta[2] = Theta[2] - 2*alpha*Sum2/m
        
    return Theta
```

```python
import pandas as pd
def Dados(caminho, norm = True):
    
    X = pd.read_csv(caminho, sep = '\\s+', header = None)
    
    if(norm):
        X = (X-X.min())/(X.max()-X.min()) # Min-Max,
        
    return X.values
```

```python
import matplotlib.pyplot as plt

def plotP(Theta, g, t):
    plt.scatter(t, g, s=75, color='r', edgecolor='k')
    plt.plot(t, Theta[0] + Theta[1]*t + Theta[2]*t*t)
    plt.show()
```

```python
Data = Dados('kick2.dat', norm = False)
t = Data[:,1]
g = Data[:,0]
Theta = [0,0,0]
ThetaX = gradienteDescendenteP(Theta, g, t, 0.01)
Theta
```

```python
plotP(ThetaX, g, t)
```

```python
t = Data[:,1]
g = Data[:,2]
Theta = [0,0,0]
ThetaZ = gradienteDescendenteP(Theta, g, t, 0.01)
```

```python
plotP(ThetaZ, g, t)
```
