```python
def gradienteDescendente(Theta, g, t, alpha):
    m = len(t)
    
    for _ in range(15000):
        Sum0, Sum1 = 0,0
        
        for i in range(m):
            hi = Theta[0] + Theta[1]*t[i]
            gi = g[i]
            
            temp = (hi - gi)
            
            Sum0 += temp*1
            Sum1 += temp*t[i]
                
        Theta[0] = Theta[0] - alpha*Sum0/m
        Theta[1] = Theta[1] - alpha*Sum1/m
        
    return Theta
```

```python
import pandas as pd
def Dados(caminho, norm = True):
    
    X = pd.read_csv(caminho, sep = '\\s+', header = None)
    
    if(norm):
        X = (X-X.min())/(X.max()-X.min()) # Min-Max,
        
    return X.values
```
